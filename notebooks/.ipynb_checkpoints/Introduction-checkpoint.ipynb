If you have never used Python or Jupyter Notebooks before, this introduction is for you. Jupyter Notebooks are a medium by which researchers can share code, prose, results and data visualizations all in one file. They were designed with the intent of supporting reproducible research <a href="https://eprints.soton.ac.uk/403913/1/STAL9781614996491-0087.pdf">Kluvyer et. al.</a> write,

> Some authors may describe computational methods in prose, as part of a general
description of research methods. But human language lacks the precision of code, and
reproducing such methods is not as quick or as reliable as it should be. Others provide
code separately as supplementary material, but it may be difficult for readers to cross
reference between code and prose, and there is a risk that the two become inconsistent
as the author works on them.
Notebooks—documents integrating prose, code and results—offer a way to
publish a computational method which can be readily read and replicated. 

Jupyter Notebooks allow users to program in R, Python, or Julia. There are modules that support additional languages as well. # An introductory query

```python
import snowflake.connector
import pandas as pd
import os
```

```python
def snowflake_to_df(q):
    """
    Uses environment variables to create a snowflake session. Sends the query q to the snowflake session and returns
    a pandas dataframe with the results.
    """
    ctx = snowflake.connector.connect(
    user=os.environ['snowflake_username'],
    password=os.environ['snowflake_pw'],
    account=os.environ['snowflake_account'],
    role=os.environ['snowflake_role'],
    warehouse=os.environ['snowflake_warehouse']
    )
    cs = ctx.cursor()
    cs.execute(q)
    return cs.fetch_pandas_all()
```

Here, we collect the data and display a preview.

```python
q = """
select * from 
"COVID19"."PUBLIC"."JHU_VACCINES"
"""
vaccines = snowflake_to_df(q)
vaccines.head()
```

We can subset the data and produce a meaningful plot.

```python
GA_vaccination_timeseries = vaccines[vaccines.STABBR=="GA"][['DATE','DOSES_ALLOC_TOTAL']]
GA_vaccination_timeseries.DATE = pd.to_datetime(GA_vaccination_timeseries.DATE).dt.strftime('%m/%d')
GA_vaccination_timeseries.dropna().plot(x='DATE',y='DOSES_ALLOC_TOTAL')
plt.title("Total COVID-19 Vaccine Doses in Georgia By Day (2020-2021)")
plt.show()
```
